```python
import os
import glob
import pandas as pd
```

```python
# 경로 설정
txt_dir = r"C:\Users\DS\Desktop\fit\dataset\labels\train" 
csv_path = "lstm_exercise_data.csv"  

data_list = []

print("텍스트 파일들을 읽어와서 CSV로 합치는 중...")
```

```python
# 폴더 내 모든 txt 파일 읽기

for txt_file in glob.glob(os.path.join(txt_dir, "*.txt")):
    filename = os.path.basename(txt_file) # 예: squat_baggy_01_0000.txt
    
    # 파일 이름으로 운동 종목(정답) 판별하기
    if "leg" in filename:
        label = 0
    elif "lunge" in filename:
        label = 1
    elif "plank" in filename:
        label = 2
    elif "pushup" in filename:
        label = 3
    else:
        continue # 이름에 운동 종목이 없으면 패스!

    # 텍스트 파일 열어서 좌표 가져오기
    with open(txt_file, "r") as f:
        lines = f.readlines()
        if len(lines) == 0:
            continue # 뼈대를 못 찾아서 빈 파일이면 패스
            
        # 첫 번째 사람의 데이터만 가져오기 (문자열을 띄어쓰기 기준으로 쪼갬)
        parts = lines[0].strip().split()
        
        # 앞의 5개(클래스0, 박스좌표 4개)는 버리고, 뒤에 51개(관절좌표)만 살림
        if len(parts) >= 56:
            keypoints = [float(p) for p in parts[5:]] # 숫자로 변환
            
            # 리스트에 [파일이름, 정답(0,1,2), 관절1x, 관절1y ...] 순서로 저장
            row = [filename, label] + keypoints
            data_list.append(row)
```

```python
# 엑셀(CSV) 파일로 저장하기
# 컬럼(맨 윗줄) 이름 만들기
columns = ["filename", "label"]
for i in range(1, 18):
    columns.extend([f"x{i}", f"y{i}", f"v{i}"])

df = pd.DataFrame(data_list, columns=columns)
df.to_csv(csv_path, index=False)

print(f"🎉 성공! 총 {len(df)}프레임의 데이터가 '{csv_path}'에 저장되었습니다.")
```
