```python
from ultralytics import YOLO
import pandas as pd
import torch
from config import DATA_YAML_PATH, RUNS_DIR

DATA_YAML = DATA_YAML_PATH

MODELS = {
    "YOLOv8s_base": "yolov8s-pose.pt",
    "YOLOv8s_finetuned": RUNS_DIR / "pose" / "runs" / "yolov8s_pose_cf-2" / "weights" / "best.pt",
    "YOLOv8x_base": "yolov8x-pose.pt"
}

device = 0 if torch.cuda.is_available() else "cpu"

results = []
```

```python

for name, model_path in MODELS.items():

    print(f"\n평가 중: {name}")

    model = YOLO(model_path)

    metrics = model.val(
        data=DATA_YAML,
        imgsz=640,
        device=device,
        verbose=False
    )

    row = {
        "Model": name,
        "Box_mAP50": round(metrics.box.map50, 4),
        "Box_mAP50_95": round(metrics.box.map, 4),
        "Pose_mAP50": round(metrics.pose.map50, 4),
        "Pose_mAP50_95": round(metrics.pose.map, 4),
        "Inference_ms": round(metrics.speed["inference"], 3)
    }

    results.append(row)

df = pd.DataFrame(results)

print("\n==========================")
print(df)
print("==========================")

df.to_csv(
    "pose_model_comparison.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\n저장 완료: pose_model_comparison.csv")
```
